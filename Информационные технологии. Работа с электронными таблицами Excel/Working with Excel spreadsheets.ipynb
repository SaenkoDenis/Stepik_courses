# 1. Модуль 1. Начальные сведения о работе с электронными таблицами

В MS Excel указание на отдельную ячейку таблицы, составленное из обозначения <br>
столбца и номера строки называют…

Выберите один вариант из списка

- [ ] результатом вычисления
- [ ] формулой введенной в ячейку
- [ ] диапазоном
- [x] адресом ячейки

---

В MS Excel диапазон ячеек А, В, С, D, Е, в строке 7 можно обозначить:

Выберите один вариант из списка

- [ ] 7А:7Е
- [ ] А7Е7
- [x] А7:Е7
- [ ] АЕ7

Для того, чтобы вставить новый столбец нужно:

Выберите один вариант из списка

- [x] Выделить столбец, левее которого надо вставить новый, и вызвать команду Главная - группа Ячейки – команда Вставить - Столбцы
- [ ] Выделить столбец, левее которого надо вставить новый, и вызвать команду Главная - группа Буфер обмена – команда Вставить - Столбцы
- [ ] Выделить два столбца, между которыми надо вставить новый, и вызвать команду Главная - группа Ячейки – команда Вставить - Столбцы

Какие имена можно давать листам электронной таблицы?

Выберите один вариант из списка

- [ ] Любое, обязательно буквы и цифры и не более 256 символов
- [ ] Любые, по желанию пользователя
- [x] Любое содержащее не более 31 символа и не содержащее символы / \ * ? : ' [ ]
- [ ] Имя листа должно иметь вид "Лист*" где вместо звездочки должно стоять число
